```python
%matplotlib qt
from neuronol_signalprocessing import *
from PyEMD import EMD
```

```python
# Import data
x = np.load('/Users/creator/analysis/methods/HilbertTransform/exampledata.npy')
x = x[7000:9001]
t = np.arange(0, len(x)*.001, .001)
```

```python
c, r = emd(x, t, plot_emd=True)
```

```python
x_recon = np.sum(c, axis=0) + r
err_huang = np.max(np.abs(x - x_recon))
print('Error in reconstruction = ', err_huang)
plt.figure()
# plt.plot(t, x, t, x_recon)
plt.plot(t, np.abs(x - x_recon))
plt.show()
```

```python
emd_py = EMD()
imfs = emd_py(x)
```
