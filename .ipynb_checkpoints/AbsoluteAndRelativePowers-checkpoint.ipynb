```python
%matplotlib qt
import mne
import pickle
import numpy as np
import os.path as op
import matplotlib.pyplot as plt
from scipy import signal
```

```python
def __detrend(data, detrend_type='linear', bp=0):
    """
    Remove linear trend along second axis from 2D array-like object `data`.
    """
    axis = -1
    if detrend_type not in ['linear', 'l', 'constant', 'c']:
        raise ValueError("Trend type must be 'linear' or 'constant'.")  
    data = np.asarray(data)
    if detrend_type in ['constant', 'c']:
        data_detrended = data - np.mean(data, axis, keepdims=True)
        return data_detrended
    else:
        dshape = data.shape
        N = dshape[axis]
        bp = np.sort(np.unique(np.r_[0, bp, N]))
        if np.any(bp > N):
            raise ValueError("Breakpoints must be less than length "
                             "of data along given axis.")
        Nreg = len(bp) - 1 
        # Find leastsq fit and remove it for each piece
        newdata = data.T
        for m in range(Nreg):
            Npts = bp[m + 1] - bp[m]
            A = np.ones((Npts, 2))
            A[:, 0] = np.arange(1, Npts+1) / Npts
            sl = slice(bp[m], bp[m+1])
            coef, _, _, _ = np.linalg.lstsq(A, newdata[sl], rcond=None)
            newdata[sl] = newdata[sl] - np.dot(A, coef)
        data_detrended = newdata.T
        
        return data_detrended
    
def __baseline_correction(data, times, baseline=(None, 0)):
    '''
    Apply baseline correction to `data`.
    '''
    bmin, bmax = baseline
    if bmin is None:
        imin = 0
    else:
        imin = np.where(times >= bmin)[0]
        imin = int(imin[0])
    if bmax is None:
        imax = len(times)
    else:
        imax = np.where(times <= bmax)[0]
        imax = int(imax[-1]) + 1

    baseline = np.mean(data[..., imin:imax], axis=-1, keepdims=True)
    corrected_data = data - baseline

    return corrected_data

def _integrate_psd(psd, f, band):
    '''
    Integrate power spectrum over desired frequency band.
    '''
    inds = [(np.abs(f - band[0])).argmin(), (np.abs(f - band[1])).argmin()]
    return np.trapz(psd[inds], f[inds])

def _computeFFTPower(X, Fs, times=None, detrend=None, baseline=None, window=None,
                     plot_power_spectrum=None):
    '''
    Compute the Fourier power spectrum of the N-channel block `X`
    [nChannels, nSamples] sampled at frequency `Fs`.
    '''
    
    # Detrend data
    if detrend and detrend is not None:
        X = __detrend(X)
        
    if baseline is not None:
        # Baseline correction
        X = __baseline_correction(X, times, baseline=baseline)
        # Discard baseline data
        qmin = np.where(times <= 0)[0]
        qmin = int(qmin[-1]) + 1
        X = X[:, qmin:]
    
    # Window size
    n = np.size(X, 1)
    
    # Windowing
    if window and window is not None:
        if window == 'hamming':
            w = np.hamming(n)
        elif window == 'hanning':
            w = np.hanning(n)
        elif window == 'cosine':
            w = signal.tukey(n)
        X = w * X
    
    # FFT using Numpy
    Xhat = np.fft.rfft(X)
    power = np.real(Xhat * np.conj(Xhat)) # FFT power spectrum
    freq = Fs / n * np.arange(n)
    k = np.arange(1, np.floor(n / 2), dtype='int') # only plot the first half of frequencies
    freq = freq[k]; power = power[:, k]; power_mean = np.mean(power, axis=0)
    
    # Plot power spectrum
    if plot_power_spectrum and plot_power_spectrum is not None:
        plt.figure()
        plt.plot(freq, power_mean, label='Numpy')
        plt.xlabel('Frequency [Hz]')
        plt.ylabel('Power spectrum')
        plt.show()
        
    return power, freq

def load_bipolar_connections(fname_bipolar_connections):
    with open(fname_bipolar_connections, 'rb') as f:
        return pickle.load(f)
```

```python
# Read preprocessed EEG data
fmax = 170 # Hz
data_path = '~/data/sample'
raw_file = op.join(data_path, 'sample_audvis_eeg.fif')
raw = mne.io.read_raw_fif(raw_file)
raw.crop(tmin=216, tmax=276) # pseudo-resting data interval
```

```python
params = {}
params['T_max'] = 60
params['sfreq'] = np.floor(raw.info['sfreq'])
params['size_epoch'] = 1.2
params['epoch_times'] = np.arange(-119, 601) / params['sfreq']
params['bipolar_connections'] = load_bipolar_connections('bipolar_connections.pickle')
```

# Method-1: Bauer 2001
Non-overlapping EEG epochs were detrended and cosine tapered. Then, the EEG epochs were submitted to a Fast Fourier Transform which computed the power (in $\mu V^2$) of the EEG. The power spectra were averaged over epochs.

Estimates of power within the frequency bands were expressed as a proportion of total power across the entire frequency range, i.e. Relative Power (RP). These relative power estimates for each frequency band were averaged across all the electrode sites.

<ins>Note</ins>: No reference is specified.

__Frequency band(s)__ = [19.5 - 39.78]
<!-- [1.56 - 3.12]  [3.90 - 7.02]  [7.80 - 12.48]  [13.26 - 18.72]   -->

__Measurement(s)__ = $AP_{mean, all channels}$

__Total measurement(s)__ = 1 band $\times$ 1 measurement = 1

__Total feature(s)__ = __Total measurement(s)__ = 1

```python
def power_M1(params, raw, plot_epoch_power_spectrum=None):
    '''
    Bauer2001
    All EEG epochs are detrended and cosine tapered. EEG epochs with zero-overlap are submitted
    to a Fast Fourier Transform which computes the power (in μV^2) of the EEG. The power spectra
    are averaged over epochs.
    '''
    # Unpack parameters
    T_max = params['T_max']
    print('A total of %d seconds of raw EEG data has been studied.' % T_max)
    Fs = params['sfreq']
    print('Fs =', Fs)
    T_ep = params['size_epoch']
    print('Each epoch has %0.1f seconds of EEG data.' % T_ep)
    times = params['epoch_times']
    print('Shape of times =', np.shape(times))
    
    # Get raw data
    raw = raw.copy()
    raw.del_proj(0) # delete Average EEG reference
    raw_data = raw.get_data(picks='eeg')
    raw_data = raw_data[:, :int(Fs * T_max)]
    print('Shape of raw_data =', np.shape(raw_data))
    
    # Data segmentation and power spectrum calculation
    n_epochs = int(np.floor(T_max / T_ep))
    print('Total number of epochs studied =', n_epochs)
    epochs = []; powers = []
    for i_ep in range(n_epochs):
        # Segment data into epochs
        start, stop = int(i_ep * np.floor(T_ep * Fs)), int((i_ep + 1) * np.floor(T_ep * Fs))
        ep_data = raw_data[:, start:stop]
        epochs.append(ep_data)
        # Compute power spectrum
        powers_ep, freqs = _computeFFTPower(ep_data, Fs, detrend=True, window='cosine',
                                         plot_power_spectrum=plot_epoch_power_spectrum)
        powers.append(powers_ep * 1e12) # in μV^2
    print('Shape of epochs =', np.shape(epochs))
    print('Shape of powers =', np.shape(powers))
    powers_mean_epochs = np.mean(powers, axis=0) # Average powers over epochs
    
    return powers_mean_epochs, freqs

def marker_bandAbsolutePower(powers, freqs, band, plot_power_spectrum=None):
    '''
    Calculate absolute power markers as used by Bauer 2001.
    The power spectra are averaged over channels. Estimates of power within a frequency band
    are estimated by integrating the averaged power spectra over the corresponding frequency
    band intervals.
    '''
    # Plot FFT
    mean_power = np.mean(powers, axis=0)
    if plot_power_spectrum and plot_power_spectrum is not None:
        _, ax = plt.subplots()
        ax.plot(freqs, mean_power, color='k')
        ax.set(title='FFT Power Spectrum', xlabel='Frequency (Hz)',
               ylabel='Power Spectrum')
        plt.show()
    
    # Integrate PSD in the desired frequency band
    ap_band = _integrate_psd(mean_power, freqs, band)
    
    return ap_band
```

```python
powers_mean_epochs, freqs = power_M1(params, raw)
ap_theta = marker_bandAbsolutePower(powers_mean_epochs, freqs, [4, 8], plot_power_spectrum=True)
ap_theta
```

# Method-2: Rangaswamy et al. 2002
The data were first transformed into bipolar derivations. Spectral densities were calculated using Fourier transform on overlapping epochs (50% overlap) with Hamming windowing.

The resulting spectral densities were aggregated into bands, divided by the bandwidth, and subsequently averaged across epochs to get the absolute power data. A logarithmic transformation of the values was applied to this data to normalize their distribution.

__Frequency band(s)__

All loci: [12.5 - 20.0]

Frontal: [20.5 - 28.0]

__Measurement(s)__ = $LOG (\frac{AP_{density, band}}{f_{band, max} - f_{band, min}})$

__Total measurement(s)__ = 2 channel-band groups $\times$ 1 measurement = 2

__Total feature(s)__ = __Total measurement(s)__ = 2

```python
def power_M2(params, raw, plot_epoch_power_spectrum=None):
    '''
    Rangaswamy2002
    The referencing is changed from monopolar to bipolar. Fourier transform is
    applied to overlapping epochs (overlap = 50%) after Hamming windowing. The
    resulting spectral powers are averaged across epochs.
    '''
    # Unpack parameters
    T_max = params['T_max']
    print('A total of %d seconds of raw EEG data has been studied.' % T_max)
    Fs = params['sfreq']
    print('Fs =', Fs)
    T_ep = params['size_epoch']
    print('Each epoch has %0.1f seconds of EEG data.' % T_ep)
    times = params['epoch_times']
    print('Shape of times =', np.shape(times))
    
    # Make a copy of raw data and delete projector for Average EEG reference 
    raw = raw.copy()
    raw.del_proj(0) # delete Average EEG reference
    
    # Rename EEG channels
    channel_renaming_dict = {name: name.replace('EEG 0', '') for name in raw.ch_names}
    _ = raw.rename_channels(channel_renaming_dict) # happens in-place
    
    # Get anodes and cathodes of bipolar montage
    bipolar_cnx = params['bipolar_connections']
    anode = []; cathode = []
    for a, c in bipolar_cnx.values():
        anode.append(a); cathode.append(c)
    
    # Set referencing as bipolar
    raw.load_data()
    raw_bip_ref = mne.set_bipolar_reference(raw, anode=anode, cathode=cathode)
    
    # Get raw data
    raw_data = raw_bip_ref.get_data(picks='eeg')
    raw_data = raw_data[:, :int(Fs * T_max)]
    print('Shape of raw_data =', np.shape(raw_data))
    
    # Data segmentation and power spectrum calculation
    n_win = int(np.floor(T_max / T_ep))
    n_epochs = int(1 + 2 * (n_win - 1))
    print('Total number of epochs studied =', n_epochs)
    epochs = []; powers = []
    for i_ep in range(n_epochs):
        # Segment data into epochs
        start, stop = int(i_ep / 2 * np.floor(T_ep*Fs)), int((i_ep / 2 + 1) * np.floor(T_ep*Fs))
        ep_data = raw_data[:, start:stop]
        epochs.append(ep_data)
        # Compute power spectrum
        powers_ep, freqs = _computeFFTPower(ep_data, Fs, detrend=False, window='hamming',
                                           plot_power_spectrum=plot_epoch_power_spectrum)
        powers.append(powers_ep * 1e12) # in μV^2
    print('Shape of epochs =', np.shape(epochs))
    print('Shape of powers =', np.shape(powers))
    powers_mean_epochs = np.mean(powers, axis=0)
    
    return powers_mean_epochs, freqs
```

```python
powers_mean_epochs, freqs = power_M2(params, raw)
ap_theta = marker_bandAbsolutePower(powers_mean_epochs, freqs, [4, 8], plot_power_spectrum=True)
ap_theta
```

# Method-3: Winterer et al. 1998
Non-overlapping consecutive EEG epochs referenced to auricular activity were extracted and submitted to spectral analysis using the Fast Fourier Transform. 

The square root of absolute power (magnitude), expressed in $\mu V$, as well as the relative power, was computed for all the frequency bands.

<ins>Note</ins>: The sample dataset used does not mention the reference used. __Ensure auricular referencing__ when working with the actual dataset for the study.

__Frequency band(s)__

[0.5 - 5.5]  [6.0 - 8.0]  [8.5 - 12.0]  [12.5 - 18.0]  [21.0 - 30.0]

__Measurement(s)__ = $\sqrt{AP}$; RP; TP

__Total measurement(s)__ = n_channels $\times$ 5 bands $\times$ 3 measurements

__Total feature(s)__ = __Total measurement(s)__

```python
def power_M3(params, raw, plot_epoch_power_spectrum=None):
    '''
    Winterer1998
    EEG epochs are submitted to a Fast Fourier Transform which computes the power spectra of
    the EEG. The power spectra are averaged over epochs.
    '''
    # Unpack parameters
    T_max = params['T_max']
    print('A total of %d seconds of raw EEG data has been studied.' % T_max)
    Fs = params['sfreq']
    print('Fs =', Fs)
    T_ep = params['size_epoch']
    print('Each epoch has %0.1f seconds of EEG data.' % T_ep)
    times = params['epoch_times']
    print('Shape of times =', np.shape(times))
    
    # Get raw data
    raw = raw.copy()
    raw.del_proj(0) # delete Average EEG reference
    raw_data = raw.get_data(picks='eeg')
    raw_data = raw_data[:, :int(Fs * T_max)]
    print('Shape of raw_data =', np.shape(raw_data))
    
    # Data segmentation and power spectrum calculation
    n_epochs = int(np.floor(T_max / T_ep))
    print('Total number of epochs studied =', n_epochs)
    epochs = []; powers = []
    for i_ep in range(n_epochs):
        # Segment data into epochs
        start, stop = int(i_ep * np.floor(T_ep * Fs)), int((i_ep + 1) * np.floor(T_ep * Fs))
        ep_data = raw_data[:, start:stop]
        epochs.append(ep_data)
        # Compute power spectrum
        powers_ep, freqs = _computeFFTPower(ep_data, Fs, detrend=False, window=False,
                                         plot_power_spectrum=plot_epoch_power_spectrum)
        powers.append(powers_ep * 1e12) # in μV^2
    print('Shape of epochs =', np.shape(epochs))
    print('Shape of powers =', np.shape(powers))
    powers_mean_epochs = np.mean(powers, axis=0) # Average powers over epochs
    
    return powers_mean_epochs, freqs
```

```python
powers_mean_epochs, freqs = power_M3(params, raw)
ap_theta = marker_bandAbsolutePower(powers_mean_epochs, freqs, [4, 8], plot_power_spectrum=True)
ap_theta
```

# Method-4: Saletu-Zyhlarz et al. 2004
EEG referenced to the average of mastoids was recorded. Spectral analysis was performed on non-overlapping epochs using the Fast Fourier Transform. The spectral curves were averaged over epochs and channels.

<ins>Note</ins>: 

1) The sample dataset used does not mention the reference used. __Ensure mastoid referencing__ when working with the actual dataset for the study.

2) This method is exactly the same as Method-3 (Winterer et al. 1998) except for mastoid/auricular referencing.

__Frequency band(s)__

[1.3 - 3.5]  [3.5 - 7.5]  [7.5 - 10.5]  [10.5 - 13]  [13 - 16]  [16 - 20]  [20 - 25]  [25 - 30]  [30 - 35]

$\delta + \theta$: [1.3 - 7.5]

$\alpha$: [7.5 - 13]

$\beta$: [13 - 35]

__Measurement(s)__

TP; -- [1]

AP/RP in 12 frequency bands; -- [$2 \times 12 = 24$]

Dominant Frequency (DF, in Hz); -- [1]

AP/RP of DF; -- [$2 \times 1 = 2$]

centre-of-gravity frequencies (centroids) of {$\delta + \theta$, $\alpha$, $\beta$, full-range}; -- [$1 \times 4 = 4$]

standard deviation of centroids of {$\delta + \theta$, $\alpha$, $\beta$, full-range}  -- [$1 \times 4 = 4$]

__Total measurement(s)__ = 1 + 24 + 1 + 2 + 4 + 4 = 36

__Total feature(s)__ = __Total measurement(s)__ = 36

```python
def power_M4(params, raw, plot_epoch_power_spectrum=None):
    '''
    Saletu-Zyhlarz2004
    EEG epochs are submitted to a Fast Fourier Transform which computes the power spectra of
    the EEG. The power spectra are averaged over epochs.
    '''
    # Unpack parameters
    T_max = params['T_max']
    print('A total of %d seconds of raw EEG data has been studied.' % T_max)
    Fs = params['sfreq']
    print('Fs =', Fs)
    T_ep = params['size_epoch']
    print('Each epoch has %0.1f seconds of EEG data.' % T_ep)
    times = params['epoch_times']
    print('Shape of times =', np.shape(times))
    
    # Get raw data
    raw = raw.copy()
    raw.del_proj(0) # delete Average EEG reference
    raw_data = raw.get_data(picks='eeg')
    raw_data = raw_data[:, :int(Fs * T_max)]
    print('Shape of raw_data =', np.shape(raw_data))
    
    # Data segmentation and power spectrum calculation
    n_epochs = int(np.floor(T_max / T_ep))
    print('Total number of epochs studied =', n_epochs)
    epochs = []; powers = []
    for i_ep in range(n_epochs):
        # Segment data into epochs
        start, stop = int(i_ep * np.floor(T_ep * Fs)), int((i_ep + 1) * np.floor(T_ep * Fs))
        ep_data = raw_data[:, start:stop]
        epochs.append(ep_data)
        # Compute power spectrum
        powers_ep, freqs = _computeFFTPower(ep_data, Fs, detrend=False, window=False,
                                         plot_power_spectrum=plot_epoch_power_spectrum)
        powers.append(powers_ep * 1e12) # in μV^2
    print('Shape of epochs =', np.shape(epochs))
    print('Shape of powers =', np.shape(powers))
    powers_mean_epochs = np.mean(powers, axis=0) # Average powers over epochs
    
    return powers_mean_epochs, freqs
```

# Method-5: Coutin-Churchman et al. 2003, 2006
EEG activity was recorded in the resting state and the data were referenced to the average of the two auricular reference electrodes. Power spectra from non-overlapping EEG epochs were computed using the Persyst EEG analysis software.

<ins>Note</ins>:

1) Since they used a proprietary software with closed-source methods, we do not know the exact method used for the estimation of power spectra. Therefore, we will use a well-functioning AP calculation method to calulate power and use it to extract all the features/biomarkers highlighted in their study. __Ensure non-overlapping epochs__ in the chosen AP method.

2) The sample dataset used does not mention the reference used. __Ensure auricular referencing__ when working with the actual dataset for the study.

AP and RP of AUD patients were compared with a normative database (composed of data from control subjects) using Z-scores. The measures were logarithmically transformed to achieve gaussianity before estimating the mean and standard deviation used for Z-transforms.

$$
Z_e = \frac{ln(P_e) - \bar{x_e}}{sd_e}
$$

$P_e$ is the given power measure at the e recording electrode site, $\bar{x_e}$ and $sd_e$ the corresponding mean and standard deviation of the same measure (after log-transform) at electrode e, obtained from the control data. An abnormal increase for a given absolute and/or relative power measure was defined as a Z score greater than 3, and an abnormal decrease as a Z score less than 3, in at least two neighboring sites for a given band.

AP power was given precedence over RP power to assign abnormalities. RP alterations without AP alterations were only considered at the affected band(s) only if both AP and RP in other bands were within normal limits.

__Frequency band(s)__

(2006) -> [0.5 - 3.0]  [3.5 - 7.5]  [8.0 - 13.0]  [14.0 - 25.0]

(2003) -> [0.5 - 3.5]  [4.0 - 7.5]  [8.0 - 13.0]  [13.5 - 25.0]

__Measurement(s)__ = AP; RP

__Total measurement(s)__ = n_channels $\times$ 7 bands $\times$ 2 measurements

__Total feature(s)__ = 1 (per subject) $\times$ 7 bands $\times$ 1 (abnormal AP | RP at 2 or more electrode sites) = 7

```python
def power_M5(params, raw, plot_epoch_power_spectrum=None):
    '''
    Default2022
    All EEG epochs are detrended and hanning windowed. EEG epochs with zero-overlap are submitted
    to a Fast Fourier Transform which computes the power (in μV^2) of the EEG. The power spectra
    are averaged over epochs.
    '''
    # Unpack parameters
    T_max = params['T_max']
    print('A total of %d seconds of raw EEG data has been studied.' % T_max)
    Fs = params['sfreq']
    print('Fs =', Fs)
    T_ep = params['size_epoch']
    print('Each epoch has %0.1f seconds of EEG data.' % T_ep)
    times = params['epoch_times']
    print('Shape of times =', np.shape(times))
    
    # Get raw data
    raw = raw.copy()
    raw.del_proj(0) # delete Average EEG reference
    raw_data = raw.get_data(picks='eeg')
    raw_data = raw_data[:, :int(Fs * T_max)]
    print('Shape of raw_data =', np.shape(raw_data))
    
    # Data segmentation and power spectrum calculation
    n_epochs = int(np.floor(T_max / T_ep))
    print('Total number of epochs studied =', n_epochs)
    epochs = []; powers = []
    for i_ep in range(n_epochs):
        # Segment data into epochs
        start, stop = int(i_ep * np.floor(T_ep * Fs)), int((i_ep + 1) * np.floor(T_ep * Fs))
        ep_data = raw_data[:, start:stop]
        epochs.append(ep_data)
        # Compute power spectrum
        powers_ep, freqs = _computeFFTPower(ep_data, Fs, detrend=True, window='cosine',
                                         plot_power_spectrum=plot_epoch_power_spectrum)
        powers.append(powers_ep * 1e12) # in μV^2
    print('Shape of epochs =', np.shape(epochs))
    print('Shape of powers =', np.shape(powers))
    powers_mean_epochs = np.mean(powers, axis=0) # Average powers over epochs
    
    return powers_mean_epochs, freqs
```

```python
powers_mean_epochs, freqs = power_M5(params, raw)
ap_theta = marker_bandAbsolutePower(powers_mean_epochs, freqs, [4, 8], plot_power_spectrum=True)
ap_theta
```
